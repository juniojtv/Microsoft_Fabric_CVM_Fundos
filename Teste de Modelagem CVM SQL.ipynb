```sql
SELECT * FROM dim_fundo
LIMIT 10;
```

```sql
-- Testando minima e maxima data da tabela
SELECT * FROM    
    (
    SELECT
        CNPJ_FUNDO,
        DT_COMPETENCIA,
        ROW_NUMBER() OVER( ORDER BY DT_COMPETENCIA ASC) AS RN_ASC,
        ROW_NUMBER() OVER( ORDER BY DT_COMPETENCIA DESC) AS RN_DESC
    FROM fact_performancediaria
    ) AS SUB
WHERE SUB.RN_ASC = 1 OR RN_DESC = 1;
```

```sql
-- Testando JOIN
SELECT 
    F.*,
    D.CNPJ_FUNDO,
    D.GESTOR,
    D.CATEGORIA
FROM fact_performancediaria F
LEFT JOIN dim_fundo D ON D.CNPJ_FUNDO = F.CNPJ_FUNDO
LIMIT 10;
```
